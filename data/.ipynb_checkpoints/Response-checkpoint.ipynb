```python
def plot_corpus_summary(corpus):
    fig = plt.figure(figsize=(20,10))
    ax1 = fig.add_subplot(231)
    ax2 = fig.add_subplot(232)
    ax3 = fig.add_subplot(233)
    
    all_docs = [str(doc).split(" ") for doc in corpus]
    all_docs = list(itertools.chain.from_iterable(all_docs))
    words_per_doc = [len(str(doc).split(" ")) for doc in corpus]
    unique_words,count_words = np.unique(all_docs,return_counts=True)
    indices_mas_usados = np.argpartition(count_words, len(count_words) - 10)[-10:]
    palabras_mas_usadas,count_palabras_mas_usadas = unique_words[indices_mas_usados],count_words[indices_mas_usados]
    
    print("Número de comentarios: {}".format(len(corpus)))
    print("Cantidad total de palabras: {}".format(len(all_docs)))
    print("Cantidad de palabras diferentes: {}".format(unique_words.size))
    print("Cantidad de palabras por comentario (Mediana): {}".format(np.median(words_per_doc)))
    
    print("Warning: Boxplot no está considerando outliers, esto es sólo para una mejor visualización\n")
    sns.histplot(words_per_doc,color = 'b', ax=ax1)
    sns.boxplot([x for x in count_words if (x>5)],orient='v',showfliers=False,color = 'c', ax=ax2)
    for item in ax3.get_xticklabels():
        item.set_rotation(45)
    sns.barplot(x=palabras_mas_usadas,y=count_palabras_mas_usadas)
    ax1.set_title("Largo de comentarios")
    ax2.set_title("Frecuencia de palabras") 
    ax3.set_title("10 palabras más usadas")
    
    
   
    
plot_corpus_summary(df.text)
```

## Creación de columnas extras

    * % Mayúsculas
    * Len(Comentario)

```python
def upper_chars(s):
    #s = s.replace('NAME','')
    alph= s.split(' ')
    
    alph = list(filter(str.isalpha, s))  # ['D', ..., 'O', 'W', ..., 'N', 'l', 'o', ...]
    if len(alph) != 0:
        return sum(map(str.isupper, alph)) / len(alph)
    else:
        print(s)
        return 0

df['uppers'] = df.text.apply(upper_chars)
df['largo'] = df.text.str.len()
df.head()
```

## Limpieza

    * to_lower
    * eliminación de stopwords
    * eliminación de caracteres raros
    * lemmatización

```python
STOPLIST = sorted(stopwords.words('english'))
STOPLIST.append('name') # palabra propia del dataset
PUNCT_DICT = {ord(punc): None for punc in punctuation if punc not in ['!',':',')','(']} #me quedo con estos
PUNCT_DICT[ord('!')] = ' !' # para considerarlo un token aparte
#PUNCT_DICT
```

```python
RE_D = re.compile('\d')
def has_numbers(string):
    # Revisa si un string contiene números
    # Noté que había muchas palabras muy contaminadas con números, las eliminé
    return RE_D.search(string)

def convert_emojis(text):
    # Convierte emojis a texto
    for emot in UNICODE_EMOJI:
        text = text.replace(emot, ' '+" ".join(UNICODE_EMOJI[emot].replace(","," ").replace(":"," ").split('_')))
    return text

nlp = spacy.load("en_core_web_sm")
#words = []

def processing(doc,stop_words=None):
    if not stop_words:
        stop_words = set([])
    doc = str(doc)
    doc = convert_emojis(doc) # convertimos los emojis
    
    
    clean_doc = doc.translate(PUNCT_DICT) # Tratamos la puntuación
    
    doc = nlp(clean_doc) # lo pasamos por spacy
    
    tokens = [token.lemma_.lower() for token in doc] # lemmatización y minúscula
    tokens = [t for t in tokens if not(has_numbers(t))] # eliminamos palabras contaminadas por números
    tokens = ['haha' if 'haha' in t else t for t in tokens] #estandarizo las risas
    return ' '.join(w for w in tokens if w not in stop_words) #elimino stopwords
    
df['processed_comments'] = df.apply(lambda x : processing(x['text']
                                                          ,stop_words=STOPLIST), axis=1)
```

Sobre la estandarización de las risas ('haha'): Noté que en redes sociales ocurre mucho que se repiten substrings ('hahahahaha', por ejemplo), (againnnnnn es otro ejemplo). Me hubiera gustado estandarizar todos los substrings repetidos, pero no alcancé a explorarlo. ### Modelo y Pipeline

```python
df.head()
```

```python
y = df['emotion']
df_train = df.drop(['emotion'],axis=1)

X_train,X_test,y_train,y_test = train_test_split(df_train,
                                                 y,random_state = 0, 
                                                 shuffle = True,test_size = 0.25)



# Oversampling no mejoró los resultados, intentaría con uno especializado. No lo voy a incluir en el proceso.
#y_oversampling =np.array([elem.split(',')[0] for elem in y_train.values]).astype(str)
#ros = RandomOverSampler(random_state=42,sampling_strategy = dicc)
#X_train,y_oversampled = ros.fit_resample(X_train,y_oversampling)
y_oversampled = y_train
y_train = pd.Series(y_oversampled).str.get_dummies(sep=",")
y_train = y_train.reindex(sorted(y_train.columns, key=lambda x: int(x)), axis=1)

y_test = pd.Series(y_test).str.get_dummies(sep=",")
y_test = y_test.reindex(sorted(y_test.columns, key=lambda x: int(x)), axis=1)

X_train.head()
```

```python
def cleaning(raw_df):
    raw_df['uppers'] = raw_df.text.apply(upper_chars)
    raw_df['largo'] = raw_df.text.str.len()
    raw_df['processed_comments'] = raw_df.apply(lambda x : processing(x['text']
                                                                              ,stop_words=STOPLIST), axis=1
                                               
                           )
    return raw_df.drop('text',axis=1)
```

```python
X_train,X_test = cleaning(X_train), cleaning(X_test)
```

#### Voy a evaluar con F1-Score Macro por el desbalanceo de clases.

* Estoy asumiendo que todas las clases son igual de importantes para el cliente

```python
# Borré varios parámetros del GS por motivos de tiempo. 
from sklearn.multiclass import OneVsRestClassifier

time_start = time.time()

preprocessor = ColumnTransformer(transformers = [('tfidf',TfidfVectorizer(), 'processed_comments'),
                                                ('scaler',StandardScaler(), ['uppers','largo'])])


pipe1  = pl.Pipeline(steps=[('pre', preprocessor),
                            ('sgd', OneVsRestClassifier(clf(random_state = 1)))])
param_grid1 = {
    'pre__tfidf__max_df':[0.6], #no consideramos palabras demasiado comunes (stopwords)
    'pre__tfidf__max_features': [1000,3000], #limitamos dimensionalidad
    'pre__tfidf__ngram_range': [(1,2)], #unigramas y bigramas
    'sgd__estimator__n_estimators': [50,100],
}

grid1 = GridSearchCV(pipe1, param_grid=param_grid1, cv=2, n_jobs=-1,verbose = 0)
grid1.fit(X_train, y_train)
y_pred = grid1.predict(X_test)
print("train time", time.strftime('%H:%M:%S', time.gmtime(time.time()-time_start)))
print(classification_report(y_test,y_pred))
print(grid1.best_params_)
```

#### Transformamos las predicciones al formato necesario

* Heurística: Si el modelo no está seguro de la clase (todas < 0.5) considero que el comentario es neutral, asigno la clase 27 (mayoritaria).

```python
#transformamos 
def transform_preds(preds):
    predicciones = []
    for elem in preds:
        pred_final = ''
        for cont in range(len(elem)):
            if elem[cont] == 1:
                if pred_final != '':
                    pred_final += ','
                pred_final += f'{cont}'
        if pred_final == '':
            pred_final = '27'
        predicciones.append(pred_final)
                    
    return pd.Series(predicciones)
preds_final = transform_preds(y_pred)
```

```python
preds_final_onehot = preds_final.str.get_dummies(sep=",")
preds_final_onehot = preds_final_onehot.reindex(sorted(preds_final_onehot.columns, key=lambda x: int(x)), axis=1)

print(classification_report(y_test,preds_final_onehot))
```
